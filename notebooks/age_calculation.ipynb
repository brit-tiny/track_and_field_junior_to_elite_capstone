```python
# Where the data is coming from and it's new home
input_dir = Path("../../../data/interim/interim_m_high_jump")

output_dir = Path("../../../data/interim/m_high_jump_with_age")
output_dir.mkdir(parents=True, exist_ok=True)

print(f"Looking for CSV files in: {input_dir.resolve()}")
print(f"New files saved to: {output_dir.resolve()}\n")

# Age calculation function
def age_calculation(DOB, meet_date):
    age = meet_date.year - DOB.year
    if (meet_date.month, meet_date.day) < (DOB.month, DOB.day):
        age -= 1
    return age


# Loop through each .csv file 
csv_files = list(input_dir.glob("*.csv"))
if not csv_files:
    # Prints which df is being processed
    print("No .csv file found")
else:
    print(f"Found {len(csv_files)} CSV files(s): {[f.name for f in csv_files]}\n")

for csv_file in csv_files:
    print(f"Processing: {csv_file.name}")
    df = pd.read_csv(csv_file)

    # Triple check fields exist
    if "DOB" not in df.columns or "Date" not in df.columns:
        print(f" Skipping {csv_file.name} - missing columns DOB or Date")
        continue

    # Correct datetime format
    df["DOB"] = pd.to_datetime(df["DOB"], format="%d %b %Y", errors="coerce")
    df["Date"] = pd.to_datetime(df["Date"], format="%d %b %Y", errors="coerce")

    # Calculate the age
    df["Age"] = df.apply(
    lambda row: age_calculation(row["DOB"], row["Date"]), axis=1)
    

    df["Age"] = (
        df["Age"]
        .round(0)
        .astype("Int64")
    )    

    # Save new file with Age column
    save_path = output_dir / csv_file.name
    df.to_csv(save_path, index=False)


    print(f"Saved with Age column: {save_path}\n")

print("All files processed.")
```

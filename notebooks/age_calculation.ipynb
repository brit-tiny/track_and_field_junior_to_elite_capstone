# Calculator Test

This notebook was used to test the age calcuation logic

```python
import pandas as pd
from datetime import datetime, date 
from pathlib import Path
```

```python

def age_calculation(DOB, meet_date=None):
    age = meet_date.year - DOB.year

    # If the meet takes place prior to the birthday the age will need to be adjusted
    if (meet_date.month, meet_date.day) < (DOB.month, DOB.day):
        age -= 1
    return age
# test with fellow alumni Rick NOJI DOB: 22 OCT 1967 Meet: Long Beach, CA (USA) 01 FEB 1992
DOB = date (1967, 10, 22)
meet_date = date (1992, 2, 1)
age_at_meet = age_calculation(DOB, meet_date)
print(f"Rick Noji's age at {meet_date} Long Beach Meet: {age_at_meet}")
        
```

### To streamline the cleaning process all high jump .csv data files are missing data in the WIND and Unnamed: 7 columns. 

```python
raw_dir = [
    Path("..//data/raw/m_high_jump"),
    Path("..//data/raw/w_high_jump")
]

clean_dir = Path("../data/interim")
clean_dir.mkdir(parents=True, exist_ok=True)

# apply to all files in the folder

for folder in raw_dir:
    for csv_file in folder.glob("*.csv"):
        print(f"Cleaning {csv_file.name}")
        df = pd.read_csv(csv_file)
        df = df.rename(columns={'Unnamed: 5': 'country'})
        df = df.drop(columns=['WIND', 'Unnamed: 7'], errors='ignore')
        save_path = clean_dir / csv_file.name
        df.to_csv(save_path, index=False)
        print(f"Saved cleaned file to: {save_path}\n")
    
print("All files cleaned and saved to data/interim/")
```
